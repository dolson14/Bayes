# Problem 3.2

As in the previous problem, I will use a data set of my own creation to do this problem.


```python
boundaries = np.array([0,50,100,150,200,400])
data = np.random.randint(5,35,size=5)
data
```

### Part A
First, we histogram the grouped data.

```python
widths = boundaries[1:]-boundaries[:-1]
plt.bar(boundaries[:-1],data/widths,width=widths,color="#7777aa")
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("Number",fontsize=18)
plt.ylabel("Frequency",fontsize=18)
```

### Part B

Second, we draw the cumulative frequency, marking the quartiles.

```python
CF = np.cumsum(data)
CF.resize(len(CF)+1)
CF = np.roll(CF,1)
plt.plot(boundaries,CF,'-s',linewidth=3,label="CF")
plt.plot([0,400],0.25*sum(data)*np.ones(2),'--',label="Q1")
plt.plot([0,400],0.5*sum(data)*np.ones(2),'--',label="Median")
plt.plot([0,400],0.75*sum(data)*np.ones(2),'--',label="Q3")
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("Number",fontsize=18)
plt.ylabel("Cumulative Frequency",fontsize=18)
plt.legend(loc='lower right')
```
